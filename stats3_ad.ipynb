#### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.



#### Probability Mass Function (PMF)

The Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value. It is used for discrete random variables, which are variables that take on a countable number of distinct values.

##### Example:
Consider a fair six-sided die. The random variable \( X \) represents the outcome of rolling the die. The PMF of \( X \) is given by:

\[
P(X = x) = \begin{cases} 
\frac{1}{6} & \text{for } x \in \{1, 2, 3, 4, 5, 6\} \\
0 & \text{otherwise}
\end{cases}
\]

This means that each of the six outcomes (1 through 6) has a probability of \( \frac{1}{6} \).

#### Probability Density Function (PDF)

The Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable to take on a particular value. It is used for continuous random variables, which are variables that can take on an infinite number of values within a given range. The PDF itself is not a probability but rather a density; to get the probability, we must integrate the PDF over an interval.

##### Example:
Consider a continuous random variable \( Y \) that follows a normal distribution with mean \( \mu = 0 \) and standard deviation \( \sigma = 1 \). The PDF of \( Y \) is given by:

\[
f_Y(y) = \frac{1}{\sqrt{2\pi}} e^{-\frac{y^2}{2}}
\]

This function describes the bell-shaped curve of the normal distribution, where the total area under the curve is 1.

##### Visual Representation:



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# PMF Example: Discrete Uniform Distribution (Fair Die)
x = np.arange(1, 7)
pmf = np.ones_like(x) / 6

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title('PMF of Rolling a Fair Die')
plt.xlabel('Outcome')
plt.ylabel('Probability')

# PDF Example: Standard Normal Distribution
y = np.linspace(-4, 4, 1000)
pdf = norm.pdf(y, 0, 1)

plt.subplot( 1, 2, 2)
plt.plot(y, pdf)
plt.title('PDF of Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
````

#### Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?


#### Cumulative Density Function (CDF)

The Cumulative Density Function (CDF) of a random variable \(X\) is a function that gives the probability that the variable \(X\) takes on a value less than or equal to \(x\). It applies to both discrete and continuous random variables.

For a random variable \(X\), the CDF \(F(x)\) is defined as:

\[
$F(x) = P(X \leq x)$
\]

The CDF provides a way to understand the distribution of probabilities across all possible values of the random variable.

#### Example:

Consider a continuous random variable \(Y\) that follows a standard normal distribution with mean \(\mu = 0\) and standard deviation \(\sigma = 1\). The CDF of \(Y\) is given by:

\[
$F_Y(y) = P(Y \leq y) = \int_{-\infty}^{y} \frac{1}{\sqrt{2\pi}} e^{-\frac{t^2}{2}} \, dt$
\]

This integral does not have a closed-form solution, so it is often computed using numerical methods or looked up in standard normal distribution tables.

#### Why is CDF Used?

1. **Probability Calculations:** The CDF provides the probability that a random variable is less than or equal to a certain value. This is useful for calculating probabilities over intervals.

2. **Comparing Distributions:** The CDF can be used to compare different distributions. If two random variables have similar CDFs, their distributions are similar.

3. **Quantiles and Percentiles:** The CDF is used to find quantiles and percentiles, which are values below which a certain percentage of data falls. For example, the median is the 50th percentile, which can be found using the CDF.

4. **Inverse CDF:** The inverse of the CDF, known as the quantile function, is used in statistical applications to generate random samples from a given distribution.

#### Visual Representation:



````python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# CDF Example: Standard Normal Distribution
x = np.linspace(-4, 4, 1000)
cdf = norm.cdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, cdf, label='CDF of Standard Normal Distribution')

plt.title('Cumulative Density Function (CDF)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()
plt.show()
````

#### Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.



##### Examples of Situations Using Normal Distribution

The normal distribution is widely used as a model in various fields due to its properties and the Central Limit Theorem, which states that the sum of many independent and identically distributed random variables tends toward a normal distribution, regardless of the original distribution. Here are some examples:

1. **Height of People:**
   - The heights of people within a specific population (e.g., adult men or women) tend to follow a normal distribution, with most individuals having a height around the average and fewer individuals being much taller or shorter.

2. **Measurement Errors:**
   - Measurement errors in scientific experiments often follow a normal distribution, assuming the errors are random and independent.

3. **IQ Scores:**
   - Intelligence Quotient (IQ) scores are designed to follow a normal distribution, with a mean of 100 and a standard deviation of 15.

4. **Test Scores:**
   - Standardized test scores, such as SAT or GRE scores, often follow a normal distribution, allowing for comparison across large groups of test-takers.

5. **Stock Prices:**
   - Daily returns of stock prices can be modeled using a normal distribution, which helps in financial modeling and risk assessment.

6. **Blood Pressure:**
   - Systolic and diastolic blood pressures of a healthy population typically follow a normal distribution.

#### Parameters of the Normal Distribution

The normal distribution is characterized by two parameters: the mean (\(\mu\)) and the standard deviation (\(\sigma\)). These parameters determine the shape and location of the distribution.

### Mean (\(\mu\))

- The mean is the central point of the distribution and represents the average value of the random variable.
- It determines the location of the peak of the bell curve.

##### Standard Deviation (\(\sigma\))

- The standard deviation measures the spread or dispersion of the distribution.
- It determines the width of the bell curve.
- A smaller \(\sigma\) results in a narrower, sharper peak, indicating that the data points are close to the mean.
- A larger \(\sigma\) results in a wider, flatter curve, indicating that the data points are more spread out from the mean.

#### Visual Representation



````python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu1, sigma1 = 0, 1   # Standard normal distribution
mu2, sigma2 = 0, 2   # Wider normal distribution
mu3, sigma3 = 2, 1   # Shifted normal distribution

# Generate data
x = np.linspace(-10, 10, 1000)
pdf1 = norm.pdf(x, mu1, sigma1)
pdf2 = norm.pdf(x, mu2, sigma2)
pdf3 = norm.pdf(x, mu3, sigma3)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, pdf1, label='$\mu=0, \sigma=1$ (Standard Normal)')
plt.plot(x, pdf2, label='$\mu=0, \sigma=2$ (Wider Distribution)')
plt.plot(x, pdf3, label='$\mu=2, \sigma=1$ (Shifted Distribution)')
plt.title('Normal Distributions with Different Parameters')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
````

#### Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

#### Bernoulli Distribution

The Bernoulli distribution is a discrete probability distribution that represents the outcome of a single Bernoulli trial, which is a random experiment with exactly two possible outcomes: "success" and "failure". It is named after the Swiss mathematician Jacob Bernoulli.

#### Example:

Consider a single toss of a fair coin, where "heads" is considered a success and "tails" a failure. The Bernoulli random variable \(X\) can be defined as:

\[
X = \begin{cases} 
1 & \text{if the coin lands heads (success)} \\
0 & \text{if the coin lands tails (failure)}
\end{cases}
\]

The probability mass function (PMF) of the Bernoulli distribution is given by:

\[
P(X = x) = \begin{cases} 
p & \text{if } x = 1 \\
1-p & \text{if } x = 0
\end{cases}
\]

Where \(p\) is the probability of success (e.g., \(p = 0.5\) for a fair coin).

#### Difference between Bernoulli Distribution and Binomial Distribution

1. **Number of Trials:**
   - Bernoulli Distribution: Represents a single trial or experiment with two possible outcomes.
   - Binomial Distribution: Represents the number of successes in a fixed number of independent Bernoulli trials.

2. **Number of Outcomes:**
   - Bernoulli Distribution: Has two possible outcomes (success or failure).
   - Binomial Distribution: Can have multiple possible outcomes (number of successes from 0 to the total number of trials).

3. **Probability Mass Function (PMF):**
   - Bernoulli Distribution: \(P(X = x) = p^x (1-p)^{1-x}\) for \(x \in \{0, 1\}\).
   - Binomial Distribution: \(P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}\) for \(k \in \{0, 1, 2, ..., n\}\), where \(n\) is the number of trials and \(p\) is the probability of success in each trial.

4. **Mean and Variance:**
   - Bernoulli Distribution: Mean (\(\mu\)) = \(p\), Variance (\(\sigma^2\)) = \(p(1-p)\).
   - Binomial Distribution: Mean (\(\mu\)) = \(np\), Variance (\(\sigma^2\)) = \(np(1-p)\), where \(n\) is the number of trials.

##### Example:

For example, consider flipping a fair coin 5 times. The Binomial distribution would represent the number of heads obtained in these 5 flips, while the Bernoulli distribution would represent the outcome of a single flip (success if heads, failure if tails).

The main difference is that the Bernoulli distribution is used for a single trial, while the Binomial distribution is used for multiple trials (repeated Bernoulli trials) with a fixed number of trials and the same probability of success for each trial.

#### Q7: Explain uniform Distribution with an example.

#### Uniform Distribution

The uniform distribution is a probability distribution where all outcomes are equally likely. In other words, each value in the distribution has the same probability of occurring. The uniform distribution is often represented as \( U(a, b) \), where \( a \) and \( b \) are the minimum and maximum values of the distribution, respectively.

##### Example

Consider a fair six-sided die. When you roll the die, each face (1, 2, 3, 4, 5, or 6) has an equal probability of \(\frac{1}{6}\) of appearing. This is an example of a discrete uniform distribution, where the outcomes are discrete (specific numbers).

Another example is the continuous uniform distribution, which is commonly used to model situations where all outcomes between a minimum and maximum value are equally likely. For example, if you were to randomly select a point from a line segment of length 10 units, and each point along the line segment has an equal chance of being selected, this would follow a continuous uniform distribution over the interval [0, 10].

##### Mathematical Representation

The probability density function (PDF) of the continuous uniform distribution is given by:

\[
f(x) = \begin{cases} 
\frac{1}{b-a} & \text{for } a \leq x \leq b \\
0 & \text{otherwise}
\end{cases}
\]

Where:
- \( a \) is the minimum value of the distribution
- \( b \) is the maximum value of the distribution

In the case of the example with the line segment of length 10, the PDF would be \( f(x) = \frac{1}{10} \) for \( 0 \leq x \leq 10 \), and \( f(x) = 0 \) for \( x < 0 \) or \( x > 10 \).

##### Visual Representation

Here's a visual representation of the continuous uniform distribution over the interval [0, 10]:

```python


````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 0  # Minimum value
b = 10  # Maximum value

# Generate data for the PDF
x = np.linspace(a, b, 1000)
pdf = np.ones_like(x) / (b - a)

# Plot
plt.figure(figsize=(9, 5))
plt.plot(x, pdf, label='Uniform Distribution PDF')
plt.fill_between(x, pdf, where=(x >= a) & (x <= b), alpha=0.3)
plt.title('Continuous Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
````

#### Q8: What is the z score? State the importance of the z score.

#### Z Score

The z-score, also known as standard score, is a measure of how many standard deviations a data point is from the mean of the dataset. It is calculated using the formula:

\[ z = $\frac{x - \mu}{\sigma}$ \]
Where:
- \( z \) is the z-score
- \( x \) is the individual data point
- \( $\mu$\) is the mean of the dataset
- \( $\sigma$ \) is the standard deviation of the dataset

##### Importance of Z Score

1. **Standardization:** Z-scores standardize the data, making it easier to compare different datasets. By converting data into z-scores, we can compare data from different distributions.

2. **Outlier Detection:** Z-scores help in identifying outliers in a dataset. Data points with z-scores significantly higher or lower than the mean may be considered outliers.

3. **Probability Calculation:** Z-scores are used in calculating probabilities in the standard normal distribution. By converting data to z-scores, we can use the standard normal distribution table to find probabilities.

4. **Data Analysis:** Z-scores are used in hypothesis testing and confidence interval calculations. They allow for the comparison of sample means and population means.

5. **Data Transformation:** Z-scores are often used in transforming skewed data into a more normal distribution, which can be useful for certain statistical analyses.

In summary, the z-score is a standardized measure that helps in comparing and analyzing data, identifying outliers, and calculating probabilities in the context of the normal distribution.


#### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

#### Central Limit Theorem (CLT)

The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean of any independent, random variable will be approximately normally distributed, regardless of the original distribution of the variable, as long as the sample size is sufficiently large.

##### Significance of the Central Limit Theorem

1. **Foundation of Statistical Inference:** The CLT is fundamental in statistical inference, as it allows us to make inferences about population parameters based on sample data. It forms the basis for many statistical tests and confidence intervals.

2. **Normal Approximation:** The CLT allows us to approximate the distribution of the sample mean as a normal distribution, even when the population distribution is not normal. This is crucial for making statistical inferences and calculations.

3. **Sample Size Determination:** The CLT helps in determining the sample size needed for hypothesis testing and constructing confidence intervals. It provides guidelines on how large the sample size should be for the sampling distribution to be approximately normal.

4. **Real-world Applications:** The CLT is widely applicable in various fields, such as finance, biology, social sciences, and more. It allows researchers to analyze data and draw conclusions even when the population distribution is unknown or not normally distributed.

5. **Quality Control:** In manufacturing and quality control processes, the CLT is used to monitor and control product quality. It helps in understanding the distribution of sample means and detecting any deviations from the expected values.

In summary, the Central Limit Theorem is a fundamental concept in statistics that allows us to make inferences about population parameters based on sample data, even when the population distribution is not normal. It has widespread applications in research, quality control, and decision-making processes.


#### Q10: State the assumptions of the Central Limit Theorem.

#### Assumptions of the Central Limit Theorem (CLT)

The Central Limit Theorem (CLT) has several key assumptions that must be met for it to hold:

1. **Random Sampling:** The samples must be drawn randomly from the population, meaning that each member of the population has an equal chance of being selected.

2. **Independence:** The samples must be independent of each other. This means that the outcome of one sample should not affect the outcome of another sample.

3. **Sample Size:** The sample size should be sufficiently large. While there is no strict rule for how large the sample size should be, a common guideline is that the sample size should be at least 30. However, the exact size may depend on the distribution of the population.

4. **Population Distribution:** While the population distribution does not need to be normal, it should have a well-defined mean and finite variance. In practical terms, this means that the population should not be extremely skewed or have heavy tails.

5. **Finite Variance:** The population from which the samples are drawn should have a finite variance. This ensures that the sample means have a well-defined distribution.

6. **Identical Distribution:** The samples should be drawn from populations with identical distributions. This assumption is often relaxed in practice, as the CLT can still hold approximately even when the populations are not identical.

7. **Finite Population:** In some cases, the CLT assumes that the population from which the samples are drawn is finite. However, this assumption is not always necessary and can be relaxed in many practical scenarios.

In summary, the Central Limit Theorem (CLT) relies on several key assumptions, including random sampling, independence, sufficiently large sample size, well-defined population distribution, and finite variance. These assumptions ensure that the sample means have a normal distribution, regardless of the original population distribution.
